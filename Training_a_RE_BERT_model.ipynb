```
reviews_test = open('test_data_'+APPNAME+'.txt', 'r').readlines()
reqs_exracted =  open(APPNAME+'_extracted_reqs.txt', 'r').readlines()
reqs_labeled =  open('test_label_'+APPNAME+'.txt', 'r').readlines()
```

```
import pandas as pd

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)

df = pd.DataFrame()
df['review'] = reviews_test
df['labeled (humans)'] = reqs_labeled
df['extracted (re-bert)'] = reqs_exracted
df = df.replace('\n',' ', regex=True)
df
```
